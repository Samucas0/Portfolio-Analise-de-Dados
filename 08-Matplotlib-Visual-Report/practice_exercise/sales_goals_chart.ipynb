# Goals Graphic
 Create a bar graphic showing the sales of 4 salesmen and a horizontal line representing the sales goal.


```python
# Célula 1: Imports and Data
import matplotlib.pyplot as plt

# Data for the chart
sales = [20, 12, 30, 17]
sellers = ['Seller 1', 'Seller 2', 'Seller 3', 'Seller 4']

# The goal is a single value, not a list that varies
sales_goal = 25
```

```python
# Célula 2: Plotting the Graph
# Create the figure and axes
plt.figure(figsize=(10, 6))

# Plot the bar chart for actual sales
plt.bar(sellers, sales, label='Actual Sales')

# --- The Correct Way to Plot a Goal Line ---
# Use axhline to draw a horizontal line across the entire chart
# y: the value on the y-axis where the line should be drawn
# color: sets the line color
# linestyle: sets the line style (e.g., dashed)
# label: the name that will appear in the legend
plt.axhline(y=sales_goal, color='red', linestyle='--', label='Sales Goal')

# --- Customization ---
plt.title('Sales Performance vs. Goal')
plt.ylabel('Units Sold')
plt.xlabel('Sellers')
plt.legend() # This activates the labels we defined

# Show the final plot
plt.show()
```
